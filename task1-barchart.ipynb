```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('pop.csv')
```

```python
df.head(10)
```

```python
df_alt = df.drop(['Country Code','Indicator Name','Indicator Code'], axis=1)
```

```python
df_alt
```

```python
chosen_country = 'South Africa'
sa = df_alt[df_alt['Country Name'] == chosen_country]
```

```python
sa
```

```python
years = sa.columns[1:].astype(int)
population = sa.iloc[:,1:].values.flatten()
```

```python
years
```

```python
population
```

```python
plt.figure(figsize=(10, 5))
plt.bar(years, population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.show()
```

```python
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
